```python
# import dependencies 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
# read in updated TSLA csv

file = pd.read_csv("TSLA_updated.csv", index_col=0, parse_dates=True)
tsla = pd.DataFrame(file)
tsla.head()
```

```python
# read in updated NFLX csv

data = pd.read_csv("NFLX_updated.csv", index_col=0, parse_dates=True)
nflx= pd.DataFrame(data)
nflx.head(3)
```

```python
# Tesla series for graphing

t_closes = tsla["Close"]
t_opens = tsla["Open"]
t_highs = tsla["High"]
t_lows = tsla["Low"]
```

```python
# netflix series for graphing

n_closes = nflx["Close"]
n_opens = nflx["Open"]
n_highs = nflx["High"]
n_lows = nflx["Low"]
```

```python
# TSLA  & NFLX closing prices 

plt.plot(t_closes, color="g")
plt.plot(n_closes, color="FireBrick")
plt.xlabel("Date")
# plt.ylabel("Price ($)")
plt.title("TSLA & NFLX Closing Price Comparision")
plt.show()
plt.savefig("TSLA_v_NFLX_Closes.png")
```

```python
# TSLA 4 piece

fig, ax = plt.subplots(2,2)
tsla["Open"].plot(ax=ax[0,0], color= "DarkViolet", title = "Opening Prices ($)")
tsla["Close"].plot(ax=ax[0,1], color="Crimson", title = "Closing Prices ($)")
tsla["High"].plot(ax=ax[1,0], color= "Aqua", title = "Daily Highest Price ($)")
tsla["Normalized_Change"].plot(ax=ax[1,1], color= "CornflowerBlue", title = "Normalized Change")
plt.tight_layout()
plt.savefig("TSLA_4_piece.png")
```

```python
# NFLX 4 piece

fig, ax = plt.subplots(2,2)
nflx["Open"].plot(ax=ax[0,0], color= "DarkViolet", title = "Opening Prices ($)")
nflx["Close"].plot(ax=ax[0,1], color="Crimson", title = "Closing Prices ($)")
nflx["High"].plot(ax=ax[1,0], color= "Aqua", title = "Daily Highest Price ($)")
nflx["Normalized_Change"].plot(ax=ax[1,1], color= "CornflowerBlue", title = "Normalized Change")
plt.tight_layout()
plt.savefig("NFLX_4_piece.png")
```
