```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Netflix data from may 23rd, 2020 -> may 23rd, 2021

file = pd.read_csv("NFLX_start.csv", index_col = 0, parse_dates = True)
df = pd.DataFrame(file)
df.head()
```

```python
df["Daily_Change"] = df["Close"] - df["Open"]
```

```python
df["Daily_Pct_Change"] = (df["Close"] - df["Open"])/df["Open"]*100
```

```python
df["Normalized_Change"] = df["Close"]/df["Close"].iloc[0]
```
