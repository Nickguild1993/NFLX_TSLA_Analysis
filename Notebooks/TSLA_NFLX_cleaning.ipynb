```python
# dependencies 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
TSLA = pd.read_csv("TSLA.csv", index_col = 0, parse_dates= True)
NFLX = pd.read_csv("NFLX.csv", index_col = 0, parse_dates= True)
```

```python
# import TSLA first, format it
tsla = pd.DataFrame(TSLA)
tsla.head(3)
```

```python
# format 

tsla = tsla.round(2)
tsla.sample(n=3)
```

```python
# export the rounded TSLA to csv

# tsla.to_csv(r'TSLA_updated.csv', header=True, index=True)
```

```python
# import & format the NFLX DF
nflx = pd.DataFrame(NFLX)
```

```python
nflx = nflx.round(2)
nflx.head(3)
```

```python
# export the rounded NFLX to csv

 # nflx.to_csv("NFLX_updated.csv", header= True, index=True)
```

```python
# merge on dates 

combined_df = tsla.join(nflx, lsuffix="_TSLA", rsuffix="_NFLX")
```

```python
combined_df.head(3)
```

```python
# export the combined df to a csv file

combined_df.to_csv("NFLX_TSLA.csv", header=True, index=True)
```
